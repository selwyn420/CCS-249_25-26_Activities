## Dataset

```python
doc = [
    ("Free money now!!!", "SPAM"),
    ("Hi mom, how are you?", "HAM"),
    ("Lowest price for your meds", "SPAM"),
    ("Are we still on for dinner?", "HAM"),
    ("Win a free iPhone today", "SPAM"),
    ("Let's catch up tomorrow at the office", "HAM"),
    ("Meeting at 3 PM tomorrow", "HAM"),
    ("Get 50% off, limited time!", "SPAM"),
    ("Team meeting in the office", "HAM"),
    ("Click here for prizes!", "SPAM"),
    ("Can you send the report?", "HAM")
    ]

```

## Normalized and tokenized the words

```python
def tokenize(text):
  # Convert to lowercase
  text = text.lower()
  # Remove punctuation by keeping only alphanumeric characters and spaces
  clean_text = "".join(char for char in text if char.isalnum() or char.isspace())
  return clean_text.split()
```

## 1.a. Function to collect words

```python
def generate_bag_of_words(documents):
  bag_of_words = {}
  for text, category_name in documents:
    if category_name not in bag_of_words:
      bag_of_words[category_name] = {}
    for word in tokenize(text):
      bag_of_words[category_name][word] = bag_of_words[category_name].get(word, 0) + 1
  return bag_of_words
```

## 1.a. Display the bag of words and total vocabulary size

```python
bag_of_words_result = generate_bag_of_words(doc)

def get_frequency(item):
  return item[1]

print("\n--- Bag of Words ---")
for category, words_freq in bag_of_words_result.items():
    print(f"\nCategory: {category}")
    # Sort words by frequency in descending order using a named function
    sorted_words = sorted(words_freq.items(), key=get_frequency, reverse=True)
    for word, freq in sorted_words:
        print(f"  {word}: {freq}")

# Define spam_words and ham_words from bag_of_words_result
spam_words = list(bag_of_words_result.get('SPAM', {}).keys())
ham_words = list(bag_of_words_result.get('HAM', {}).keys())

vocabulary = spam_words + ham_words
print(f"\nThe vocabulary size is: {len(vocabulary)}")
```

##1.b. Calculate the priors

```python
def calculate_priors(documents):
    total_docs = len(documents)
    category_counts = {}
    for _, category in documents:
        category_counts[category] = category_counts.get(category, 0) + 1

    priors = {category: count / total_docs for category, count in category_counts.items()}
    return priors

priors = calculate_priors(doc)
print("Prior Probabilities:")
for category, prob in priors.items():
    print(f"  P({category}) = {prob:.4f}")
```

## 1.c. Likelyhood of the tokens (with respect to the class)

```python
def calculate_likelihoods(bag_of_words, vocabulary):
    likelihoods = {}
    for category, words_freq in bag_of_words.items():
        likelihoods[category] = {}
        total_words_in_category = sum(words_freq.values())

        for word in vocabulary:
            # Apply Laplace smoothing (add-1 smoothing)
            word_count = words_freq.get(word, 0)
            # The denominator should include the size of the vocabulary for Laplace smoothing
            likelihood = (word_count + 1) / (total_words_in_category + len(vocabulary))
            likelihoods[category][word] = likelihood
    return likelihoods

# Create a vocabulary of all unique words from all categories
vocabulary = set()
for category_name, words_freq in bag_of_words_result.items():
    vocabulary.update(words_freq.keys())

likelihoods = calculate_likelihoods(bag_of_words_result, vocabulary)

def get_likelihood_value(item):
  return item[1]

print("\nLikelihoods (P(word|class)):")
for category, word_likelihoods in likelihoods.items():
    print(f"\nCategory: {category}")
    # Sort likelihoods for better readability
    sorted_likelihoods = sorted(word_likelihoods.items(), key=get_likelihood_value, reverse=True)
    for word, prob in sorted_likelihoods:
        print(f"  P({word}|{category}) = {prob:.4f}")
```

## 1.d. Test Sentences

```python
import math

# Pre-calculate total words per category for use in unseen word likelihood
total_word_counts_per_category = {
    category: sum(words_freq.values()) for category, words_freq in bag_of_words_result.items()
}

# Pre-calculate unseen word likelihoods per category
unseen_word_likelihoods_per_category = {
    category: 1 / (total_word_counts_per_category[category] + len(vocabulary))
    for category in priors.keys()
}

def classify_sentence(sentence, priors, likelihoods, vocabulary, unseen_word_likelihoods_per_category):
    tokenized_sentence = tokenize(sentence)

    log_posteriors = {category: math.log(prior) for category, prior in priors.items()}

    for category in priors.keys():
        # Use the pre-calculated unseen word likelihood value for this category
        unseen_word_likelihood_value = unseen_word_likelihoods_per_category[category]

        for word in tokenized_sentence:
            # Get likelihood for the word in the current category
            # If the word is not in our likelihoods, use the 'unseen_word_likelihood_value'.
            likelihood = likelihoods[category].get(word, unseen_word_likelihood_value)

            # Add the log likelihood to the total log posterior for this category
            if likelihood == 0:
                log_posteriors[category] += -math.inf # Highly improbable
            else:
                log_posteriors[category] += math.log(likelihood)

    # Find the class with the maximum log posterior
    predicted_class = max(log_posteriors, key=log_posteriors.get)

    return predicted_class, log_posteriors

# Test sentences
sentence1 = "Limited offer, click here!"
sentence2 = "Meeting at 2 PM with the manager."

# Classify sentence 1
predicted_class1, log_probs1 = classify_sentence(sentence1, priors, likelihoods, vocabulary, unseen_word_likelihoods_per_category)
print(f"i.) Sentence: '{sentence1}'")
print(f"Predicted Class: {predicted_class1}")
print(f"Log Probabilities: {log_probs1}")

# Classify sentence 2
predicted_class2, log_probs2 = classify_sentence(sentence2, priors, likelihoods, vocabulary, unseen_word_likelihoods_per_category)
print(f"\nii.) Sentence: '{sentence2}'")
print(f"Predicted Class: {predicted_class2}")
print(f"Log Probabilities: {log_probs2}")
```
